**Task-2 :** Predict the optimum number of clusters and represent it visually.

```python
#importing all the neccessary libraries 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
```

```python
#Reading data
iris = datasets.load_iris()
#and storing it as a dataframe 'df'
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
iris_df.shape
#rows include 50 flowers of each samples- Setosa, Versicolor and Virginica
```

```python
# head keyword displays top 5 rows of dataframe 'df'
iris_df.head(10)
```

```python
#.describe() gives statical information about columns of dataframe 'iris_df'
iris_df.describe()
```

```python
#finding optimum number of clusters
#by THE ELBOW METHOD
x = iris_df.iloc[:, [0, 1, 2, 3]].values
```

```python
print(x)
```

```python
from sklearn.cluster import KMeans
wss = []  #within-cluster-sum of squared errors

#for different number of clusters 'n_cluster'
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wss.append(kmeans.inertia_)
```

```python
#plotting graph between wss and number of clusters
plt.plot(range(1, 11), wss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WSS') 
plt.show()
#for given dataset, the elbow clear at n_cluster = 3
```

```python
# for number of clusters = 3
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

```python
# Visualising the clusters
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')
# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```
